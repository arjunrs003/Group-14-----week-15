# Data Preprocessing

```python
#to sort out numerical values
numerical_features=[feature for feature in df.columns if df[feature].dtypes!='O']
numerical_features
print('The length of numerical features is :',len(numerical_features))
```

```python
df[numerical_features].head()
```

```python
#let us find out the discrete variables
discrete_features=[feature for feature in numerical_features if len(df[feature].unique())<25]
print('Discrete features counts :',len(discrete_features))
```

```python
df[discrete_features]
```

```python
df[discrete_features].columns
```

```python
continous_features=[feature for feature in numerical_features if feature not in discrete_features]
print('Continous feature count is:',len(continous_features))
```

```python
continous_features
```

```python
continous=df[continous_features]
continous
```

```python
#to plot boxplot
for i in continous_features:
    df.boxplot(column=i)
    plt.show()
```

```python
#to find the outlier values and handling it. 
for i in continous.columns:
    for j in df["intrusion"].unique():
            classtype = df[df["intrusion"] == j]
            column = classtype[i]
            q1 = column.quantile(0.25) 
            q3 = column.quantile(0.75) 
            iqr = q3 - q1 
            upper_limit = q3 + 1.5 * iqr
            lower_limit = q1 - 1.5 * iqr        
            outliers = column[(column > upper_limit) | (column < lower_limit)].index 
            df.drop(index = outliers, inplace=True)
```

```python
df.shape
```

```python
#to sort out categorical values
categorical_features=[feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
print('The length of categorical features is :',len(categorical_features),'\nThe categorical features are ',categorical_features)
```

```python
#let us encode these features
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
df['protocol_type']=label_en.fit_transform(df['protocol_type'])
df['service']=label_en.fit_transform(df['service'])
df['flag']=label_en.fit_transform(df['flag'])
df['intrusion']=label_en.fit_transform(df['intrusion'])
```

'serror_rate','srv_serrer_rate','dst_host_srv_serror_rate' and 'dst_host_serror_rate' are highly correlated with each other so we remove the columns 'serror_rate','srv_serrer_rate'and'dst_host_srv_serror_rate'

 'rerror_rate','srv_rerror_rate','dst_host_srv_rerror_rate' and 'dst_host_rerror_rate' are highly correlated with each other so we remove the columns 'rerror_rate','srv_rerror_rate'and'dst_host_srv_rerror_rate'

'same_srv_rate','dst_host_srv_count','dst_host_same_srv_rate' are highly correlated we remove 'same_srv_rate','dst_host_srv_count'

```python
df.drop(['serror_rate','srv_serrer_rate','dst_host_srv_serror_rate','rerror_rate','srv_rerror_rate','dst_host_srv_rerror_rate','same_srv_rate','dst_host_srv_count','num_outbound_cmds','is_host_login'],inplace=True,axis=1)
```

```python
df.shape
```

```python
#splitting the dataset
x=df.drop(['intrusion'],axis=1)
y=df['intrusion']
```

```python
#let us standardize the data
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
```

```python
x=std_scaler.fit_transform(x)
```

```python
#let us split train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# Modelling

```python
# GaussianNB classifier
from sklearn.naive_bayes import GaussianNB
classifierGauss = GaussianNB()
classifierGauss.fit(x_train,y_train)
```

```python
y_pred=classifierGauss.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred, average = 'micro'))
```

```python
# Bernoulli classifier
from sklearn.naive_bayes import BernoulliNB
classifierB = BernoulliNB()
classifierB.fit(x_train,y_train)
y_pred=classifierB.predict(x_test)
```

```python
print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred, average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred, average = 'micro'))
```
